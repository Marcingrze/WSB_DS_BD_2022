```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def w1(x):
    wynik=[]
    for i in x:
        if i<=6:
            wynik.append(0)
        else:
            wynik.append(4 * (i+6)**0.5)
    return wynik
def w2(x):
    return ((3 * x**3) + (4 * x**2) - (5*x) + math.sqrt(7))
def w3(x):
    return x*x
```

```python
x=np.arange(-10,10,0.1)
```

```python
y=w1(x)
y2=w2(x)
y3=w3(x)
```

```python
plt.plot(x,y,x,y2, x, y3)
plt.legend(["w1", "w2", "w3"], loc=1)
plt.show()
#plt.plot(x,y2/100,"r")
```

```python
plt.plot(x,y,"r")
```

```python
plt.subplot(121)
plt.plot(x,y3,"g")
plt.xlabel("x")
plt.ylabel("y2(x)")
plt.title("w2")

plt.subplot(122)
plt.plot(x,y2,"r")
plt.title("w1")
plt.xlabel("x")
plt.ylabel("y(x)")

plt.suptitle("Zadanie z wyk≈Çadu")
plt.show()
```
